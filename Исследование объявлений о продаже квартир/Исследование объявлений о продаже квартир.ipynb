На данном этапе необходимо избавиться от выбросов. Создадим функцию, определяющую границы "усов" для задаваемого признака.

```python
def det_whiskers(column):
    stat = data[column].describe()
    
    # межквартильный размах
    iqr = stat[6] - stat[4]
    
    # левый и правый "ус"    
    left_whisker = round(stat[4] - 1.5 * iqr, 2)
    right_whisker = round(stat[6] + 1.5 * iqr, 2)

    # левый "ус" не должен быть меньше минимального значения
    if left_whisker < stat[3]: left_whisker = stat[3] 
        
    # правый "ус" не должен быть больше максимального значения
    if right_whisker > stat[7]: right_whisker = stat[7]
        
    return [left_whisker, right_whisker]
```

```python
#зададим функцию, вычисляющую долю аномальных значений
def anomalies_ratio(column):
    return 'Аномальные значения составляют {:.1%}'. format(len(data.loc[data[column] > det_whiskers(column)[1]]) / len(data))
anomalies_ratio('total_area')
```

Изучим зависимость полной цены от типа этажа.

```python
#Перепишем типы этажей в цифры, чтобы посчитать корреляцию
def new_new_floor(row):
    try:
        if row['floor'] == 1:
            return 0 #Первый этаж
        if row['floor'] == row['floors_total']:
            return 2 #Последний этаж
        return 1 #Другой этаж
    except:
        return 'нет информации'
data['new_new_floor'] = data.apply(new_new_floor, axis = 1)  
    
```

```python
#Посмотрим на коэффициент корреляции 
data['new_new_floor'].corr(data['last_price'])
```

```python
#Построим график зависимости
data.plot(x='new_new_floor', y='last_price', kind='scatter', alpha=0.5) 
plt.show();
```

**Вывод:**
    
Зависимость видна по графику, есть квартиры где, самая высокая цена - на последнем этаже. Но если брать в целом, то на "другом" распределение построено так, что в среднем оно дороже "последнего".
